# Basic graph in Matplot

```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore
import pandas as pd # type: ignore
```

```python
x = [5,10,15,20,30]
y = [4,6,8,20,12]

plt.plot(x,y)
plt.plot(x,y,color='red',marker='*')
plt.show()
```

```python
plt.plot(x,y,color='red',marker='*',linestyle='dashdot',markeredgecolor='black',markersize=10)
```

```python
#for heading of the graph
plt.plot(x,y,'r*--')
plt.title('Graph 1',fontdict={'fontname':'comic Sans Ms','fontsize':20})
```

```python
#labeling the axis and legend for the graph
plt.plot(x,y,'b*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Perfect graph 1',fontdict={'fontname':'comic Sans Ms','fontsize':20})
plt.legend(['single element'])
plt.plot(x,y,'b*--')
plt.savefig('perfectgraph.png',dpi=200)
plt.show()
```

# Line graph in Matplot

```python
import pandas as pd 

df = pd.read_csv('data 2.csv')
df.head()
```

# Plotting multiple graphs

```python
#plotting multiple graphs

plt.plot(df.year,df.india,'b.-',label='India')
plt.plot(df.year,df.china,'r.-',label='China')
plt.plot(df.year,df.america,'g.-',label='America')
plt.plot(df.year,df.japan,'m.-',label='Japan')

plt.xlabel('year')
plt.ylabel('price')
plt.title('Mobile Price Year Wise',fontdict={'fontname':'comic Sans Ms','fontsize':20})

plt.legend()
plt.savefig('mobileprice.png',dpi=300)
plt.show()
```

# Piechart Graph

```python
import numpy as np
y = np.array([25,30,35,40])
name = ["Raj","Krishna","Radha","Rupa"]

plt.pie(y,labels=name)
plt.show()
```

```python
nameexplode = [0.1,0.2,0.3,0.2]
plt.pie(y,labels=name,explode=nameexplode)
plt.show
```

```python
plt.pie(y,labels=name,explode=nameexplode,shadow=True)
plt.savefig('pie.png',dpi=100)
plt.legend()
plt.show()
```
